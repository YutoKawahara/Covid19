```python
df1 = pd.read_csv(fullpath1, error_bad_lines = False)
newrows = list(df1["Combined_Key"])
df1.index = newrows
df1.shape
```

```python
step = 10
for i in range(0, 100, step):
  UScases =  df1.iloc[i:(i+step),11:].transpose()
  UScases.plot() #first 100 rows plotted in the 10 small figures
```

```python
groupby_state = df1.iloc[:, 11:].groupby(df1['Province_State'])
df_by_state = groupby_state.sum()
df_by_state.head()
```

```python
df_by_state.transpose().plot()
```

```python
df_by_state.transpose().iloc[:,0:5].plot(subplots = True)
```

```python
df_by_state.loc['Wyoming'].plot()
```

```python
df_by_state.head()
```

```python
df_dailycases_by_state = df_by_state.diff(axis = 1) # 1 Difference by columns
df_dailycases_by_state.head()
```

```python
df_dailycases_by_state.transpose()[['Tennessee', 'Wisconsin', 'Wyoming']].plot()
```

```python
df_dailycases_by_state.transpose()[['Tennessee','Alaska']].rolling(window = 3).mean().plot(logy=True)
```

```python
df_dailycases_by_state.transpose()[['Tennessee','Alaska']].rolling(window = 3).mean().plot(subplots=True)
```
